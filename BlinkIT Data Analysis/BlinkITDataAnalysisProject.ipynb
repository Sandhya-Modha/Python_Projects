**1. Insight From The Above Plot:**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Item_Outlet_Sales'], kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Count')
plt.show()
```

**2. Histogram (Order Frequency & Spending Behavior):**

This histogram displays the distribution of order values, helping us identify common spending patterns among customers.<br>
Most orders fall within a specific value range, indicating common spending patterns.
If right-skewed, smaller orders are more frequent; if left-skewed, high-value purchases dominate.<br>
Peaks in the histogram highlight the most frequent order amounts, which can guide pricing strategies

```python
# Select only numerical columns for correlation calculation
numeric_dataset = df.select_dtypes(include=['number'])

# Now you can create the heatmap
sns.heatmap(numeric_dataset.corr(),
            cmap='BrBG',
            fmt='.2f',
            linewidths=2,
            annot=True)
```

**3. Heatmap (Correlation Between Variables)**

Highlights strong relationships between variables like order value vs. delivery time or order volume vs. day of the week.
Darker/lighter areas indicate higher/lower correlation, guiding business optimizations.<br>
A high correlation between **order value and delivery time** suggests faster deliveries for high-value customers.

```python
sns.histplot(df[df['Item_MRP']<100]['Item_MRP'])
plt.grid(axis='y')
```

**4. Distribution of Low-Cost Products (MRP < ₹100):**

This histogram shows the price distribution of products with an MRP below ₹100. It helps in identifying the most common price ranges for budget-friendly items.


```python
# Calculate sales contribution by outlet type
sales_by_outlet_type = df.groupby('Outlet_Type')['Item_Outlet_Sales'].sum()

# Create a pie chart
plt.figure(figsize=(5, 5))
myexplode=[0,0.2]
sales_by_outlet_type.plot.pie(
    autopct='%1.1f%%',  
    startangle=180,
    colors=['#ff9999', '#66b3ff','#008000','#FFFF00'],
    shadow=True,
    explode=myexplode
)
plt.title('Sales Contribution by Outlet Type')
plt.ylabel('')  # Hide y-axis label
plt.show()
```

**5. Pie Chart (Sales Contribution by Outlet Type):**

This pie chart represents the sales contribution of different outlet types.
The **largest segment** indicates the outlet type that generates the highest revenue.
If one outlet type dominates sales, it suggests **higher customer preference or better-performing locations.**
The **exploded slice** (highlighted section) draws attention to a specific outlet type that has a significant but not dominant contribution.
If contributions are evenly distributed, it indicates a balanced sales performance across outlet types.
Helps in **strategic expansion, marketing focus, and resource allocation** to high-performing outlet types.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Item_Type', order=df['Item_Type'].value_counts().index)
plt.title('Distribution of Item Types')
plt.xlabel('Count')
plt.ylabel('Item Type')
plt.show()
```

**6. Bar Chart (Popular Product Categories):**

This bar chart showcases the most frequently purchased product categories.
Identifies the most ordered product categories, helping in inventory management.<br>
A few categories may contribute the most revenue (Pareto principle: 80/20 rule).<br>
Helps in strategizing promotions and stocking decisions.<br>

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Item_MRP'], kde=True, bins=30, color='purple')
plt.title('Distribution of Item MRP')
plt.xlabel('Item MRP')
plt.ylabel('Frequency')
plt.show()
```

**7. Histogram (Distribution of the "Item MRP"):**

The distribution of the "Item MRP" will provide a clear picture of your pricing strategy—whether it's focused on low-cost items, high-end products, or a balanced mix. This insight can inform decisions on product pricing, positioning, and potential strategies for optimizing the product mix. For example, if most items are concentrated at a lower price point, there may be room to introduce higher-priced items for greater margins.
